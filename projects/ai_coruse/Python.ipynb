### 数值型

数值型用于存储数值。自Python 3.6起，数值中可加入下划线以增强可读性。支持四种不同的数字类型：

| 类型 | 名称 | 示例 | 特点 |
|------|------|------|------|
| `int` | 整型（整数） | `a = 10` | 没有小数点，可正可负 |
| `float` | 浮点型（小数） | `b = 3.14` | 带小数点的实数，遵循 IEEE 754 |
| `bool` | 布尔型 | `c = True` | 逻辑值，仅有 True / False |
| `complex` | 复数型 | `d = 1 + 2j` | 包含实部与虚部 |

特点：
- **`int`（有符号整型）**：
    - 可以是正整数或负整数，不带小数点。
    - Python 的 int 没有固定字长，可根据需要自动扩展为大整数（不同于 C 语言的 int32、int64 限制）。
- **`float`（浮点型）**：
    - 浮点型由整数部分与小数部分组成。
    - 精度问题是计算机领域的普遍现象，IEEE 754标准使用二进制表示浮点数，涉及符号位、指数位和尾数位，这使得在0.5范围内的数值四舍五入结果可能不准确。
- **`bool` （布尔型）**：
    - 描述True 和 False，类型是整型（int）的子类。
    - 需区分逻辑运算符（not、and、or）与按位运算符（按位与、按位或）。
    - 官方定义中，逻辑运算符 “x or y” 在x为真值时返回x，否则返回y；而 “not” 运算符逻辑明确，非真即假。
- **`complex`（虚数）**：
    - 复数由实数部分和虚数部分构成，可以用a + bj,或者complex(a,b)表示， 复数的实部a和虚部b都是浮点型。



````python
a = 10_000 # 10_000 是 Python 允许的写法（提高可读性）,实际值就是 10000
print(a)
b = 10.0
print(a or b)
print(not b)
````

2️⃣ 浮点型（float）

2.1 浮点数由**整数部分 + 小数部分**组成，遵循 IEEE 754 标准。  底层以二进制形式存储，十进制小数被截断存储从而产生误差，运算后误差被放大，导致出现精度误差。

````python
a = 0.5 + 0.5
print(a == 1)

b = 0.1 + 0.3
print(b == 0.4)

c = 0.1 + 0.2
print(c == 0.3) #  ​IEEE 754 标准和二进制表示法的局限性

# 一个十进制小数能被二进制浮点数精确表示，当且仅当它可以写成：\frac{k}{2^n}
````

2.2 通过格式化字符串打印浮点数的二进制表示，以观察误差

````python
print(f'{0.1:.20f}')
print(f'{0.2:.20f}')
print(f'{0.3:.20f}')
print(f'{0.4:.20f}')

print(f'{0.1 + 0.3:.20f}') 
print(f'{0.1 + 0.2:.20f}') 
````

2.3 四舍五入陷阱：

```round()```函数可以对数字进行“四舍五入”操作（rounding），可指定保留的小数位数。

````python
print(round(4.78, 1))  
print(round(4.55, 1))  # 思考为什么输出不是 4.6
print(f'{4.55:.10f}')
print(f'{4.55:.20f}')
````

🧩提示：因为底层二进制存储的精度问题，4.55 实际在内存中略小于 4.55（如 4.549999...），所以四舍五入后变成了 4.5。

````python
print(True * 100)
print(True + 100)
print(False + 10)
````

3.2 真值与假值（Truthy / Falsy）

在 Python 中，几乎所有对象都可以当作布尔值来判断真假。

所有的对象都可以直接转换为布尔值，比如数字 0、空字符串 `''`、空列表 `[]` 等都被认为是假值（Falsy）

其他所有非零数值、非空容器等都被认为是真值（Truthy）。

````python
print(bool(0), bool(3.14), bool([]), bool("hello"), bool(None))
````

3.3 逻辑运算返回值规则与短路特性（Short Circuit）

在 Python 中，```and``` 和 ```or``` 不会返回布尔值，而是:
- ```and``` 返回第一个“假值”或最后一个“真值”
- ```or``` 返回第一个“真值”或最后一个“假值”

- ```and``` 与 ```or``` 具有“短路”特性。即：一旦能确定结果，后面的表达式就不会再执行。

````python
0 and print("这不会执行")      # 短路，后面不执行
1 and print("这会执行")       # 不短路

print(0 and 4==5)        # 返回0（短路）
print(5 and 10)          # 返回10（最后一个真值）
print(34 and print(5))   # 先打印5，然后返回None
````

**元组的解包操作**

解包是Python中一种强大而优雅的语法特性，它允许我们将可迭代对象中的元素自动地、按顺序地分配给一系列对应的变量。这种特性在列表、元组、字符串等所有可迭代对象中都适用。

```*```的用法：收集其余所有的剩余元素。在变量前加上 ```*```，这个变量就会变成一个<span style="color:red">**列表**</span>。<br>
⚠️注意：解包操作并不局限于元组，它适用于几乎所有可迭代对象，如列表、字符串等

````python
coordinates = (10, 20, 30, 40, 50) # 元组
x, *y_values, z = coordinates # 解包
print(x , y_values, z)

x, *y_values, z, w = coordinates # 解包
print(x , y_values, z, w)
````

### 字典与集合

### 函数基础要点

下面介绍函数相关的几个重要概念：

- **函数的参数与返回值**：函数接收输入参数并返回处理结果的机制。
- **函数的作用域**：函数内部变量的可见性和生命周期。
- **匿名函数与高级函数**：简洁的无名称函数和可操作其他函数的函数。
- <span style="color: #999999">**闭包函数与装饰器**：闭包能保留函数状态，装饰器则是闭包的一种实用应用，可自行深入了解。</span>


1. 全局变量的作用域，在函数外部定义，函数内部可以访问（只读时）
2. 局部变量的作用域，在函数内部定义，函数外部不能访问

````python
a = 10
def test():
    b = a + 5
    print(b)
    
test()
````

修改全局变量需要global关键字

````python
a = 10

def test():
    global a
    a = 20

test()
print(a)  # 20
````

| 语句         | 作用         |
| ---------- | ---------- |

| `continue` | 跳过本次，继续下一次 |

````python
l = [1,2,3,4]
def for_loop(l): # 定义一个函数，用于遍历列表中的每一个元素
    temp = iter(l) # 创建一个迭代器对象
    while True:
        try:
            print(next(temp)) # 打印当前元素
        except StopIteration: # 当迭代器没有更多元素时，抛出StopIteration异常
            break
for_loop(l)
````

### 列表和array的速度对比
- NumPy（Numeric Python）
  - NumPy 所有元素类型需要相同（同质性），而列表元素内容可以不同，这使得列表的内存比较碎片化。
  - 列表还可以扩容，当预留的大小不够时，会分配一个更大的内存空间，并将所有内容复制过去。
  - 列表存储的是其元素的引用，而非实际的元素。每个引用指向列表元素在内存中的实际位置，
    这允许列表高效地操作不同类型的对象，并且支持对同一对象的多个引用。
- NumPy 具有以下特性：
  - 向量化运算
  - 广播机制
  - <span style="color:#CCCCCC;">内存视图（view 方式，可节省内存和提升性能，避免了数据的赋值，但可能会破坏数据）</span>


````python
import numpy as np

arr = np.array([1,2,3,4,5])
print(arr[0])
arr[0] = 66 # 将数组的第一个元素赋值为整数66
print(arr)

# NumPy 数组要求所有元素类型一致（dtype统一）,赋值时会尝试把新值转换成数组已有类型
arr[0] = '88' # 将数组的第一个元素赋值为字符串'88'
print(arr)    # 数组为什么不是['88',2,3,4,5]? 因为数组的元素是不同类型的，不能直接赋值给数组的元素，只能将元素转换为统一的类型后赋值
````

````python
arr1 = np.array([1,2,3,4,5],dtype=np.float32)   # 将数组的元素转换为浮点数类型
arr2 = np.array([2,3,4,5,6],dtype=np.float32)
arr1 + arr2 #向量化运算，避免了解释器层面的循环
````

````python
c = 100
arr1 + c #广播机制:在 NumPy 中，当两个数组形状不同但满足一定规则时，可以自动“扩展”较小的数组，使其与较大数组进行逐元素运算
````

- 类：定义属性和方法的模板，对象是类的实例

- 实例化：创建对象的过程
- 对象：类的实例，包含实例变量，并可访问类变量和方法

1️⃣类与对象基本概念
- **类（class）**：描述对象的模板（属性 + 方法）。
- **对象 / 实例（instance）**：由类创建的具体实体。

- **构造函数**：`__init__(...)` 用来初始化实例状态（属性）。
- 在 Python中，`self`是一个惯用的名称，用于表示类的实例（对象）自身。它是一个指向实例的引用，使得类的方法能够访问和操作实例的属性。
- 当你定义一个类，并在类中定义方法时，第一个参数通常被命名为`self`，尽管你可以使用其他名称，但强烈建议使用`self`，以保持代码的一致性和可读性。

````python
class MyClass:
# 类有一个名为 __init__() 的特殊方法（构造方法），在创建对象(类实例化)时自动调用，用来初始化对象属性的特殊方法
    def __init__(self):
        self.data = []
````

类定义了 `__init__(...)` 方法，类的实例化操作会自动调用 `__init__(...)` 。如下实例化类 MyClass，对应的 `__init__(...)`方法就会被调用:

````python
x = MyClass()
print(x.data)
````

当然， `__init__(...)` 方法也可以有参数，参数通过 `__init__(...)` 传递到类的实例化操作上。例如:

````python
## 1. 简单类定义
class Student:
    def __init__(self, age, sex):
        self.age = age
        self.sex = sex
    
    def speaking(self):
        print('苟利国家生死以')

    def thinking(self):
        print('我思故我在')
````

解析
- `__init__`：构造函数，在 `Student(...)` 创建实例时自动执行，通常用于给实例绑定属性 `self.age`、`self.sex`。
- `self`：实例自身的引用，用于访问或设置实例属性。
- `speaking` / `thinking`：普通实例方法，通过 `实例.方法()` 调用，隐含传入 `self`。


````python
## 2. 创建对象并调用方法

xiaoming = Student(18,'M')  # 创建Student对象
# Student.thinking(xiaoming)  # 可以直接这样写
xiaoming.thinking()         # 等价于Student.thinking(xiaoming) # 调用方法
xiaoming.speaking()
````

### 继承
Python 支持类的继承（Inheritance），用于复用代码、扩展功能、建立类之间的层次关系。

````python
#类定义
class people:
    #定义基本属性
    name = ''
    age = 0
    #定义构造方法
    def __init__(self,n,a):
        self.name = n
        self.age = a
    def speak(self):
        print("%s 说: 我 %d 岁了。" %(self.name,self.age))
 
#单继承示例
class student(people):
    grade = ''
    def __init__(self,n,a,g):
        #调用父类的构造函数
        people.__init__(self,n,a)
        self.grade = g
    #覆写父类的方法
    def speak(self):
        print("%s 说: 我 %d 岁了，我在读 %d 年级 "%(self.name,self.age,self.grade))
 
p = people('Peter',10)
p.speak()

s = student('James',10, 3)
s.speak()
````

### 方法重写
如果你的父类方法的功能不能满足你的需求，你可以在子类重写你父类的方法，实例如下：

````python
class Parent:        # 定义父类
   def myMethod(self):
      print ('调用父类方法A')
 
class Child(Parent): # 定义子类
   def myMethod(self):
      print ('调用子类方法B')
 
c = Child()          # 子类实例
c.myMethod()         # 子类调用重写方法
super(Child,c).myMethod() #用子类对象调用父类已被覆盖的方法
````

### 类属性与方法

- 类的私有属性
__private_attrs：两个下划线开头，声明该属性为私有，不能在类的外部被使用或直接访问。在类内部的方法中使用时 self.__private_attrs。


- 类的私有方法
__private_method：两个下划线开头，声明该方法为私有方法，只能在类的内部调用 ，不能在类的外部调用。self.__private_methods。


类的私有属性实例如下：

````python

class JustCounter:
    __secretCount = 0  # 私有变量
    publicCount = 0    # 公开变量
 
    def count(self):
        self.__secretCount += 1
        self.publicCount += 1
        print (self.__secretCount)
 
counter = JustCounter()
counter.count()
counter.count()
counter.count()
print (counter.publicCount)
# print (counter.__secretCount) # 报错，实例不能访问私有变量
````

类的私有方法实例如下：

````python
class Site:
    def __init__(self, name, url):
        self.name = name       # public
        self.__url = url   # private
 
    def who(self):
        print('name  : ', self.name)
        print('url : ', self.__url)
 
    def __foo(self):          # 私有方法
        print('这是私有方法')
 
    def foo(self):            # 公共方法
        print('这是公共方法')
        self.__foo()
 
x = Site('交大官网', 'www.sjtu.edu.cn')
x.who()        # 正常输出
x.foo()        # 正常输出
# x.__foo()      # 报错，实例不能访问私有方法
````

### 运算符重载 (zhong zai)
运算符重载指在自定义类中定义特殊方法，使对象可以使用内置运算符（如 + - * ==），并让这些运算具有你指定的含义。

````python
a = 10
b = 20
a + b   #   本质是调用：a.__add__(b)
a - b   #   本质是调用：a.__sub__(b)
a * b   #   本质是调用：a.__mul__(b)
a / b   #   本质是调用：a.__truediv__(b)
````

````python
# 运算符重载本质上是让类“像内置类型一样自然使用”
# 示例：自定义“向量相加”
class Vector:
   def __init__(self, a, b):
      self.a = a
      self.b = b
 
   def __str__(self):
      return 'Vector (%d, %d)' % (self.a, self.b)
   
   def __add__(self,other):
      return Vector(self.a + other.a, self.b + other.b)
 
v1 = Vector(2,10)
v2 = Vector(5,-2)
print (v1 + v2) # Python 实际执行的是 v1.__add__(v2)
````
